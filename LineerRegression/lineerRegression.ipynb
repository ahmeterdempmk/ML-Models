```python
import pandas as pd

# Load dataset
df = pd.read_csv('C:\\Users\\SAMSUNG\\Desktop\\Makine Öğrenmesi\\Makine Öğrenmesi 201\\Basit Doğrusal Regresyon\\Advertising.csv')
df = df.iloc[:, 1:len(df)]
df.head()

import seaborn as sns
sns.jointplot(x = "TV", y = "sales", data = df, kind = "reg")

X = df[["TV"]] # Independent variable
y = df[["sales"]] # Dependent variable
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = df[["TV"]] # Independent variable
y = df[["sales"]] # Dependent variable
reg = LinearRegression()
model = reg.fit(X, y) # Model setup

print("Models Beta0 coefficient: ", model.intercept_)
print("Models Beta1 coefficient: ", model.coef_)

# rsquared
print("Modelin RSquared value: ", model.score(X, y)) # It is the percentage of the change in the dependent variable explained by the independent variables.
```

```python
# Visual representation
g = sns.regplot(x = "TV", y= "sales", data = df, ci=None, scatter_kws={'color': 'r', 's':9})
g.set_title("Model Equation: Sales = 7.03 + TV*0.05")
g.set_ylabel("Number of Sales")
g.set_xlabel("TV Expenses")
plt.xlim(-10, 310)
plt.ylim(bottom=0)
```
